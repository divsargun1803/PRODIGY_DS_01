## 3. Exploratory Data Analysis (EDA)

## 4. Visualizations

```python
# ── Theme & Colors ──────────────────────────────────────────────────────────
BG      = '#0D1117'
CARD    = '#161B22'
TEXT    = '#E6EDF3'
SUBTEXT = '#8B949E'
GRID    = '#21262D'
ACCENT1 = '#58A6FF'
ACCENT2 = '#F78166'
ACCENT3 = '#3FB950'
ACCENT4 = '#D2A8FF'

REGION_COLORS = {
    'Sub-Saharan Africa':       '#F78166',
    'South Asia':               '#58A6FF',
    'East Asia':                '#3FB950',
    'Southeast Asia':           '#D2A8FF',
    'Latin America':            '#FFA657',
    'Europe':                   '#79C0FF',
    'Middle East & North Africa':'#FF7B72',
    'North America':            '#56D364',
    'Central Asia':             '#E3B341',
}
INCOME_COLORS = {
    'High':          '#3FB950',
    'Upper middle':  '#58A6FF',
    'Lower middle':  '#FFA657',
    'Low':           '#F78166',
}

plt.rcParams.update({
    'figure.facecolor': BG, 'axes.facecolor': CARD,
    'axes.edgecolor': GRID, 'axes.labelcolor': TEXT,
    'xtick.color': SUBTEXT, 'ytick.color': SUBTEXT,
    'text.color': TEXT, 'grid.color': GRID, 'grid.linewidth': 0.5,
    'axes.spines.top': False, 'axes.spines.right': False,
    'axes.spines.left': False, 'axes.spines.bottom': False,
})
print('Theme applied ✅')
```

### 4.1 Top 15 Most Populous Countries — Horizontal Bar Chart

```python
top15 = df.nlargest(15, 'Population_2022').sort_values('Population_2022')
colors_bar = [REGION_COLORS[r] for r in top15['Region']]

fig, ax = plt.subplots(figsize=(14, 7), facecolor=BG)
bars = ax.barh(top15['Country'], top15['Population_2022']/1e9, color=colors_bar, height=0.65, zorder=3)
for bar, val in zip(bars, top15['Population_2022']/1e9):
    ax.text(val + 0.01, bar.get_y() + bar.get_height()/2,
            f'{val:.2f}B', va='center', fontsize=9, color=TEXT)

ax.axvline(x=1, color=SUBTEXT, linewidth=1, linestyle='--', alpha=0.5)
ax.text(1.01, 0.5, '1B', transform=ax.get_xaxis_transform(), fontsize=8, color=SUBTEXT)
ax.set_xlabel('Population (Billions)', fontsize=10, color=SUBTEXT, labelpad=8)
ax.set_title('Top 15 Most Populous Countries (2022)', fontsize=14, fontweight='bold', color=TEXT, pad=12, loc='left')
ax.grid(axis='x', zorder=0)

seen = {}
for c, r in zip(colors_bar, top15['Region']):
    if r not in seen: seen[r] = mpatches.Patch(color=c, label=r)
ax.legend(handles=list(seen.values()), loc='lower right', framealpha=0.15,
          fontsize=9, ncol=2, facecolor=CARD, edgecolor=GRID)

plt.tight_layout()
plt.savefig('chart1_top15_countries.png', dpi=150, bbox_inches='tight', facecolor=BG)
plt.show()
print('Chart 1 saved ✅')
```

### 4.2 Population by Region — Bar Chart (Categorical Variable)

```python
region_pop = df.groupby('Region')['Population_2022'].sum().sort_values(ascending=False) / 1e9
rc = [REGION_COLORS[r] for r in region_pop.index]

fig, ax = plt.subplots(figsize=(13, 6), facecolor=BG)
b = ax.bar(range(len(region_pop)), region_pop.values, color=rc, width=0.65, zorder=3)
for bar, val in zip(b, region_pop.values):
    ax.text(bar.get_x() + bar.get_width()/2, val + 0.03,
            f'{val:.2f}B', ha='center', fontsize=9, color=TEXT)

ax.set_xticks(range(len(region_pop)))
ax.set_xticklabels(
    [r.replace(' & ', '\n& ').replace(' ', '\n', 1) if len(r) > 14 else r
     for r in region_pop.index], fontsize=8.5
)
ax.set_ylabel('Total Population (Billions)', fontsize=10, color=SUBTEXT)
ax.set_title('Total Population by Region (2022)', fontsize=14, fontweight='bold', color=TEXT, pad=12, loc='left')
ax.grid(axis='y', zorder=0)
plt.tight_layout()
plt.savefig('chart2_region.png', dpi=150, bbox_inches='tight', facecolor=BG)
plt.show()
print('Chart 2 saved ✅')
```

### 4.3 Population by Income Group — Bar Chart

```python
income_pop = df.groupby('Income_Group')['Population_2022'].sum().sort_values(ascending=False) / 1e9
ic = [INCOME_COLORS[i] for i in income_pop.index]

fig, ax = plt.subplots(figsize=(8, 5), facecolor=BG)
b = ax.bar(income_pop.index, income_pop.values, color=ic, width=0.55, zorder=3)
for bar, val in zip(b, income_pop.values):
    ax.text(bar.get_x() + bar.get_width()/2, val + 0.05,
            f'{val:.2f}B', ha='center', fontsize=10, color=TEXT)

ax.set_ylabel('Total Population (Billions)', fontsize=10, color=SUBTEXT)
ax.set_title('Population Distribution by Income Group (2022)', fontsize=13, fontweight='bold', color=TEXT, pad=12, loc='left')
ax.grid(axis='y', zorder=0)
plt.tight_layout()
plt.savefig('chart3_income.png', dpi=150, bbox_inches='tight', facecolor=BG)
plt.show()
print('Chart 3 saved ✅')
```

### 4.4 Histogram — Distribution of Country Populations (Continuous Variable)

```python
pop_mill = df['Population_2022'] / 1e6

fig, ax = plt.subplots(figsize=(13, 5), facecolor=BG)
n, bins, patches = ax.hist(pop_mill, bins=15, color=ACCENT1, edgecolor=BG, linewidth=0.8, zorder=3)

# Color bins by population magnitude for readability
for patch, left in zip(patches, bins[:-1]):
    if left < 100:
        patch.set_facecolor(ACCENT1)
    elif left < 300:
        patch.set_facecolor(ACCENT4)
    else:
        patch.set_facecolor(ACCENT2)

ax.axvline(pop_mill.median(), color=ACCENT3, linewidth=2, linestyle='--',
           label=f'Median: {pop_mill.median():.0f}M')
ax.axvline(pop_mill.mean(), color=ACCENT2, linewidth=2, linestyle='--',
           label=f'Mean: {pop_mill.mean():.0f}M')

ax.set_xlabel('Population (Millions)', fontsize=11, color=SUBTEXT, labelpad=8)
ax.set_ylabel('Number of Countries', fontsize=11, color=SUBTEXT)
ax.set_title('Histogram: Distribution of Country Populations (2022)',
             fontsize=14, fontweight='bold', color=TEXT, pad=12, loc='left')
ax.legend(fontsize=10, framealpha=0.15, facecolor=CARD, edgecolor=GRID)
ax.grid(axis='y', zorder=0)
ax.annotate('Most countries\nhave <200M people', xy=(80, 20), fontsize=9,
            color=SUBTEXT, ha='center')

plt.tight_layout()
plt.savefig('chart4_histogram.png', dpi=150, bbox_inches='tight', facecolor=BG)
plt.show()
print('Chart 4 saved ✅')
```

## 5. Key Insights

1. **Highly skewed distribution** — Most countries have populations below 200 million; only China and India exceed 1 billion, making the histogram strongly right-skewed.

2. **Asia dominates** — South Asia and East Asia together account for nearly half the world's population across just a handful of countries.

3. **Sub-Saharan Africa is fragmented** — The region has the most countries in the sample, yet its total population is still smaller than South Asia's, reflecting smaller per-country averages.

4. **Income & population mismatch** — Lower-middle income countries contain the largest total population, highlighting the development challenge: the most people live in nations with the least resources per capita.

5. **Median vs Mean gap** — The large gap between median (~52M) and mean (~159M) is driven by the extreme outliers (China, India, USA), confirming population is not normally distributed.
